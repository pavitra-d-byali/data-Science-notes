# Geometric Intuition Behind Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is used for **dimensionality reduction**.  
It helps us extract new features from the original features while retaining as much information (variance) as possible.

---

## Example Setup

Suppose we have a housing dataset:

- Features:
  - `size_of_house` (X1)
  - `number_of_rooms` (X2)
- Output:
  - `price` (Y)

We want to reduce **2D features → 1D feature** while preserving maximum information.

---

## Simple Projection Approach

1. **Direct Projection on X-axis:**  
   - Project all points onto one feature (e.g., `size_of_house`)  
   - Result: 2D → 1D  
   - Problem: Information about `number_of_rooms` is lost  

2. **Observation:**  
   - **Spread** of points = **variance**  
   - Larger spread → more variance  
   - Direct projection loses variance from the second feature  

**Conclusion:** Simple projection is a form of feature extraction, but **loses significant information**.

---

## PCA Transformation

1. PCA applies a **linear transformation** to the original axes.  
   - Uses **eigendecomposition** on the covariance matrix of the data  
   - Creates **new axes** (principal components)  

2. **New axes:**
   - `PC1` → Principal Component 1 (captures **maximum variance**)  
   - `PC2` → Principal Component 2 (captures **next maximum variance**, perpendicular to PC1)  

3. **Projection onto new axes:**
   - Project all points onto `PC1` to reduce dimensionality  
   - Minimal information loss compared to direct projection  

---

## Key Points

- **Goal of PCA:** Find the line(s) (principal components) that capture the **maximum variance** of the data.  
- **Variance importance:** High variance indicates that the feature contains more information.  
- **Ordering of Principal Components:**
  - `PC1` → captures most variance  
  - `PC2` → captures second most variance  
  - `PC3` → captures third most variance (for 3D data)  
- **Dimensionality Reduction:**
  - 2D → 1D → project onto `PC1`  
  - 3D → 1D → project onto `PC1`  
  - 3D → 2D → project onto `PC1` and `PC2`  

---

## PCA Example Visualization

Suppose we have 2D points (X1, X2):

1. Original axes:



X1-axis → size of house
X2-axis → number of rooms


2. Apply PCA transformation:



PC1-axis → direction of maximum variance
PC2-axis → perpendicular to PC1, direction of next maximum variance


3. Project points:

- Points projected onto `PC1` capture **most variance**
- Minimal information from `PC2` is lost

---

## Summary

- PCA finds new axes (principal components) that maximize variance  
- The **first principal component (PC1)** captures the most variance  
- Additional components capture remaining variance, orthogonal to previous ones  
- Dimensionality reduction is done by projecting points onto the top k principal components  

> The final goal is to reduce dimensions while **retaining maximum information**.

# Mathematical Intuition Behind Principal Component Analysis (PCA)

In this section, we will understand the **mathematical intuition** behind PCA.

---

## Example Setup

Consider 2D points plotted along axes:

- X-axis → feature 1  
- Y-axis → feature 2  

Goal: **Reduce 2D → 1D** while capturing **maximum variance**.

- PCA finds the **best principal component (PC1)** line such that variance is maximized.

---

## Key Concepts

Two important concepts in PCA:

1. **Projections**
2. **Cost function (variance)**

---

## Projection of a Point

Let a point be:

$$
P_1 = (x_1, y_1)
$$

Consider a **unit vector** along a line (principal component):

$$
\mathbf{u} = (u_x, u_y)
$$

- Projection of \(P_1\) onto \(\mathbf{u}\) is denoted as \(P_1'\)  

### Formula for Projection

If \(\mathbf{u}\) is a unit vector:

$$
P_1' = P_1 \cdot \mathbf{u}
$$

> Here, the dot product gives a **scalar value** representing the distance along the unit vector.

- Repeat this for all points:

$$
P_2', P_3', \dots, P_n'
$$

- These scalar values represent **distances along the principal component**.

---

## Variance Computation

- Let the projected scalar values be:

$$
x_1', x_2', \dots, x_n'
$$

- Variance formula:

$$
\text{Var}(X') = \frac{1}{n} \sum_{i=1}^{n} (x_i' - \bar{x}')^2
$$

- **Goal:** Find unit vector \(\mathbf{u}\) that **maximizes variance**:

$$
\mathbf{u}_{\text{best}} = \arg\max_{\mathbf{u}} \text{Var}(X')
$$

> This becomes our **cost function** for PCA.

---

## Why Not Use Hit-and-Trial?

- Iterating over all possible unit vectors is inefficient.  
- Solution: Use **Eigenvectors and Eigenvalues** from linear algebra.

---

## Steps Using Eigenvectors

1. Compute **covariance matrix** of features:

$$
\Sigma = \text{Cov}(X)
$$

2. Solve the **eigenvalue equation**:

$$
\Sigma \mathbf{v} = \lambda \mathbf{v}
$$

- \(\mathbf{v}\) → eigenvector  
- \(\lambda\) → eigenvalue (magnitude of eigenvector, represents variance along that direction)

3. Select **eigenvector with largest eigenvalue**:

- This eigenvector is **PC1** → captures **maximum variance**  
- Second largest eigenvalue → PC2, etc.

> Mathematically, projection onto this eigenvector guarantees **maximum variance capture**.

---

## Summary

- PCA finds the **best unit vector** (principal component) such that projected data has **maximum variance**  
- Steps:
  1. Compute **covariance matrix**  
  2. Find **eigenvectors and eigenvalues**  
  3. Select **eigenvector with largest eigenvalue** → PC1  
- Projection onto PC1 reduces dimensionality with **minimal information loss**  

---

# Eigenvectors and Eigenvalues in PCA

In this video, we discuss **eigenvectors** and **eigenvalues** and their role in Principal Component Analysis (PCA).

---

## Motivation

- PCA aims to find the **best principal component line** that captures **maximum variance**.
- **Eigenvectors and eigenvalues** help us determine this best line mathematically.

---

## Linear Transformation

- Consider a vector:

$$
\mathbf{v} = \begin{bmatrix} v_x \\ v_y \end{bmatrix}
$$

- Applying a **linear transformation** with matrix \(A\):

$$
A \mathbf{v} = \lambda \mathbf{v}
$$

- Here:
  - \(\mathbf{v}\) → eigenvector
  - \(\lambda\) → eigenvalue (magnitude along \(\mathbf{v}\))

> Intuition: A linear transformation **stretches or compresses** vectors along certain directions (eigenvectors).  

- The eigenvector with **largest eigenvalue** captures **maximum variance** → becomes **PC1**.

---

## Covariance Matrix

To compute eigenvectors and eigenvalues in PCA:

1. Consider two features \(X\) and \(Y\)
2. Covariance between \(X\) and \(Y\):

$$
\text{Cov}(X,Y) = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

3. Covariance matrix for two features:

$$
\Sigma = 
\begin{bmatrix}
\text{Var}(X) & \text{Cov}(X,Y) \\
\text{Cov}(Y,X) & \text{Var}(Y)
\end{bmatrix}
$$

> Notes:
> - Diagonal elements → variances  
> - Off-diagonal elements → covariances  
> - Cov(X,Y) = Cov(Y,X)

- For 3 features (X, Y, Z), the covariance matrix is 3x3:

$$
\Sigma = 
\begin{bmatrix}
\text{Var}(X) & \text{Cov}(X,Y) & \text{Cov}(X,Z) \\
\text{Cov}(Y,X) & \text{Var}(Y) & \text{Cov}(Y,Z) \\
\text{Cov}(Z,X) & \text{Cov}(Z,Y) & \text{Var}(Z)
\end{bmatrix}
$$

---

## Eigenvalues and Eigenvectors

- Solve the eigenvalue equation:

$$
\Sigma \mathbf{v} = \lambda \mathbf{v}
$$

- \(\lambda_1, \lambda_2, \dots, \lambda_n\) → eigenvalues  
- Corresponding \(\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n\) → eigenvectors  

> Largest eigenvalue → PC1 (captures **maximum variance**)  
> Second largest → PC2, etc.

---

## PCA Steps (2D → 1D Example)

1. **Standardize data** (zero mean)  
   $$ X_{\text{std}} = X - \bar{X}, \quad Y_{\text{std}} = Y - \bar{Y} $$

2. **Compute covariance matrix**:

$$
\Sigma = 
\begin{bmatrix}
\text{Var}(X) & \text{Cov}(X,Y) \\
\text{Cov}(Y,X) & \text{Var}(Y)
\end{bmatrix}
$$

3. **Compute eigenvectors and eigenvalues**:

$$
\Sigma \mathbf{v} = \lambda \mathbf{v}
$$

4. **Select principal components**:

- PC1 → eigenvector with **largest eigenvalue**  
- PC2 → eigenvector with **second largest eigenvalue**

5. **Project data onto PC1** to reduce **2D → 1D**:

$$
X_{\text{PCA}} = X_{\text{std}} \cdot \mathbf{v}_1
$$

> If reducing 3D → 2D, combine PC1 and PC2 projections.

---

## Summary

- PCA finds **principal components** that capture **maximum variance**.  
- Steps:
  1. Standardize data  
  2. Compute covariance matrix  
  3. Compute eigenvectors and eigenvalues  
  4. Select eigenvectors with largest eigenvalues (PC1, PC2, …)  
  5. Project original data onto selected principal components

- **Eigenvectors** → directions of maximum variance  
- **Eigenvalues** → magnitude of variance captured

> Using this method, PCA reduces dimensionality **with minimal information loss**.


```python
# Import libraries
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Example dataset
# X = features, y = labels (0 or 1)
X = np.array([[2,3],[1,5],[2,8],[3,6],[4,7],[5,8],[6,7]])
y = np.array([0,0,0,1,1,1,1])

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Predicted Labels:", y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate sample dataset (2 features, 2 classes)
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1, random_state=42)

# Train logistic regression
model = LogisticRegression()
model.fit(X, y)

# Plot data points
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', edgecolor='k', s=50)

# Plot decision boundary
x_values = np.linspace(X[:,0].min(), X[:,0].max(), 200)
y_values = -(model.coef_[0][0] * x_values + model.intercept_[0]) / model.coef_[0][1]
plt.plot(x_values, y_values, label='Decision Boundary', color='green')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.legend()
plt.show()
```
