## Supervised Machine Learning

When we say **supervised machine learning algorithm**, it means the dataset given to us has:

- Some **features**: \( f_1, f_2, f_3, f_4 \)  
- One **dependent (output) feature** \( y \)  

Here:

- \( y \) is the **dependent/output feature**  
- \( f_1, f_2, f_3, f_4 \) are **independent/input features**

The problem statements in supervised learning are typically of two types:

1. **Regression** – Predicting continuous values  
2. **Classification** – Predicting discrete classes  

---

## Unsupervised Machine Learning

Now, let's discuss **unsupervised machine learning** and the difference from supervised learning.

In **unsupervised machine learning**, the goal is to **group data into clusters** (Clustering).  

Here:

- The dataset contains features (e.g., age, years of experience, salary)  
- There is **no specific output feature**  

Unlike supervised learning, we are **not predicting a dependent variable**.  

Instead, we are grouping similar data points together.

---

### Example of Clustering

Suppose the dataset contains:

| Age | Experience | Salary |
|-----|------------|--------|
| 25  | 2          | 50k    |
| 28  | 3          | 55k    |
| 35  | 10         | 90k    |
| ... | ...        | ...    |

**Goal:** Group people with **similar age, experience, and salary** into clusters.

- Cluster 1 → People with similar age, experience, and salary  
- Cluster 2 → Another group with different similarities  

**Result:** Similar data points are clustered together.  

---

### Real-World Use Case: Customer Segmentation

1. Suppose we own a product and have a list of customers with:

   - Salary  
   - Spending score  

2. By applying **unsupervised ML**, we can segment customers:

   - Group A → People who buy products immediately  
   - Group B → People who buy occasionally  

3. Based on segmentation:

   - Offer 15% discount to frequent buyers  
   - Offer 20% discount to occasional buyers  

This is a classic example of **customer segmentation** using clustering.
