# What is Streamlit?

**Streamlit** is an **open-source app framework** for **Machine Learning (ML)** and **Data Science** projects.  

 Key features:
- Create **interactive and beautiful web applications** with just **Python scripts**.
- No need to learn **HTML, CSS, or JavaScript**.  
- Extensively used for:
  - Prototyping ML apps  
  - Dashboards  
  - Generative AI chatbots
