#  Introduction to Probability

**Definition:**  
Probability measures the likelihood of an event occurring.

**Formula (display):**

$$
P(E) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes in the sample space}}
$$

---

##  Examples

### Tossing a Coin
- Sample space: $\{H, T\}$ → 2 outcomes.  
- $P(H)=\dfrac{1}{2},\quad P(T)=\dfrac{1}{2}$.

### Rolling a Die
- Sample space: $\{1,2,3,4,5,6\}$ → 6 outcomes.  
- $P(X=1)=\dfrac{1}{6},\; P(X=2)=\dfrac{1}{6}, \dots$

---

#  Mutually Exclusive Events

**Definition:** Two events are *mutually exclusive* if they cannot occur at the same time.

**Additive Rule (Mutually Exclusive):**

$$
P(A \;\text{or}\; B) = P(A) + P(B)
$$

**Examples:**
- Coin: $P(H \;\text{or}\; T)=\dfrac{1}{2}+\dfrac{1}{2}=1$.  
- Die: $P(1 \;\text{or}\; 5)=\dfrac{1}{6}+\dfrac{1}{6}=\dfrac{2}{6}=\dfrac{1}{3}$.

---

#  Non-Mutually Exclusive Events

**Definition:** Two events are *non-mutually exclusive* if they can occur at the same time.

**Additive Rule (Non-Mutually Exclusive):**

$$
P(A \;\text{or}\; B) = P(A) + P(B) - P(A \;\text{and}\; B)
$$

**Example (52-card deck):**  
- $P(K)=\dfrac{4}{52}$ (4 kings)  
- $P(\text{Heart})=\dfrac{13}{52}$ (13 hearts)  
- $P(K \cap \text{Heart})=\dfrac{1}{52}$ (King of Hearts)

So:

$$
P(K \;\text{or}\; \text{Heart}) = \dfrac{4}{52} + \dfrac{13}{52} - \dfrac{1}{52} = \dfrac{16}{52} = \dfrac{4}{13}
$$


#  Probability – Multiplication Rule
##  Independent Events

**Definition:**  
Two events are **independent** if the occurrence of one does **not** affect the occurrence of the other.

**Examples:**
- Tossing a coin twice (outcomes of the first toss don’t affect the second)  
- Rolling a die twice  

**Rule (Independent Events):**

$$
P(A \;\text{and}\; B) = P(A) \times P(B)
$$

**Example:**  
What is the probability of getting a **Head** on the first toss and a **Tail** on the second toss?

$$
P(H \;\text{and}\; T) = P(H) \times P(T) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}
$$

---

##  Dependent Events

**Definition:**  
Two events are **dependent** if the occurrence of one **does affect** the occurrence of the other.

**Example:**  
Drawing cards from a deck **without replacement**.  

- First draw: probability of King = $\tfrac{4}{52}$  
- Second draw (Queen after removing a King): probability = $\tfrac{4}{51}$  

**Rule (Dependent Events / Conditional Probability):**

$$
P(A \;\text{and}\; B) = P(A) \times P(B \mid A)
$$

Here, $P(B \mid A)$ = probability of **B given A has already occurred**.  

**Example:**  
Probability of drawing a **King** first, then a **Queen** without replacement:

$$
P(K \;\text{and}\; Q) = P(K) \times P(Q \mid K) = \frac{4}{52} \times \frac{4}{51} = \frac{16}{2652} = \frac{4}{663}
$$

---

##  Connection to Conditional Probability

The formula for dependent events is the same as **conditional probability**:

$$
P(A \;\text{and}\; B) = P(A) \times P(B \mid A)
$$

This concept is very important in **Naive Bayes** (a machine learning algorithm based on Bayes’ theorem).

---

##  Recap

1. **Independent Events:**  
   $P(A \;\text{and}\; B) = P(A) \times P(B)$  

2. **Dependent Events:**  
   $P(A \;\text{and}\; B) = P(A) \times P(B \mid A)$
